# Data Cleaning and Exploration

Start by importing stuff that we think might be useful

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import spearmanr, kstest, t as tdist, jarque_bera, norm, skew, laplace
from statsmodels.stats.diagnostic import acorr_ljungbox
import seaborn as sns
```

Let's take a look at the data. T1 and T2 are the names we will give to the two time series.

```python
data = pd.read_csv("quantTest_data.csv", header = None, names=["t", "T1", "T2"])
#Read in the times and convert from datenum to actual dates
data['Time'] = pd.to_datetime(data.t-719529, unit='D')
data.set_index('Time', inplace=True, drop=False)
data.describe()
```

The counts don't match up! This means missing data! We will have to deal with this later. But let's take a look at the data first.

```python
plt.plot((data.T1-data.T1.mean())/data.T1.std())
plt.plot((data.T2-data.T2.mean())/data.T2.std())
plt.legend(['T1', 'T2'])
plt.savefig('plot.png')
```

This is encouraging! The time-series definitely look negatively correlated. Possibly a mean reversion strategy might be able to leverage this?

Let's take a peek at the distributions of the data as well. First let's test for normality.

```python
print(f"Jarque-Bera p-value for T1 is:{jarque_bera(data['T1'].dropna()).pvalue}")
print(f"Jarque-Bera p-value for T2 is:{jarque_bera(data['T2'].dropna()).pvalue}")
```

Let's take a look at the normal fit.

```python
dataT1 = data['T1'].dropna()
mu, std = norm.fit(dataT1)
plt.hist(dataT1, bins=100, density=True, alpha=0.6, color='b')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)  
plt.plot(x, p, 'k', linewidth=2)
```

```python
dataT2 = data['T2'].dropna()
mu, std = norm.fit(dataT2)
plt.hist(dataT2, bins=100, density=True, alpha=0.6, color='b')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)  
plt.plot(x, p, 'k', linewidth=2)
```

The fits look terrible. However, this is expected since we haven't de-trended the data yet. The individual clusters we see will hopefully merge when later we take the difference and de-trend.

Now let's go back to the missing data issue. We need to understand the reason for the missing data. Hopefully all of the missing data is from the weekend and other holidays, but it's highly unlikely since the two stocks don't have the same numbers of missing data.

Let's first create some extra data columns to help us understand the missing data

```python
data["Hour"] = data.Time.apply(lambda row:row.hour)
data["Minute"] = data.Time.apply(lambda row:row.minute)
data["tdate"] = np.floor(data.t).astype(int)
data["Timedate"] = pd.to_datetime(data.tdate-719529, unit='D')

data["Year"] = data.Time.apply(lambda row:row.year)
data["Month"] = data.Time.apply(lambda row:row.month)
data["Quarter"] = data.Time.apply(lambda row:row.quarter)

data["DoY"] = data.Time.apply(lambda row:row.day_of_year)
data["DoW"] = data.Time.apply(lambda row:row.day_of_week)

data["T1_missing"] = data.T1.apply(lambda row:1 if np.isnan(row) else 0)
data["T2_missing"] = data.T2.apply(lambda row:1 if np.isnan(row) else 0)

data.head()
```

How much of the data is missing?

```python
print(f"Missing {data['T1_missing'].sum()} data points among {data['T1_missing'].count()} rows of variable T1 ({data['T1_missing'].sum()*100.0/data['T1_missing'].count():.2f}%)")
print(f"Missing {data['T2_missing'].sum()} data points among {data['T2_missing'].count()} rows of variable T2 ({data['T2_missing'].sum()*100.0/data['T2_missing'].count():.2f}%)")
```

Variable T1 has obviously gone through the ringer, but thankfully T2 seems to be more or less complete. Let's try to understand the skewness of the distribution of time spacings for the irregular data we have. According to Rehfeld et al. 2011, the skewness of the time spacing distribution might give an indication of whether an imputation method might do okay in handling the missing data. Specifically the 2011 paper focuses on errors in computing the autocorrelation function, but given that times series are (including T1 and T2 as we will see later), usually defined by their autocorrelation properties, this would be a good first heuristic to exclude extreme cases.

Let's calculate skewness for T1 first.

```python
times_T1 = data.loc[data.T1_missing==0]['Time']
dt = times_T1.diff().apply(lambda x: x.seconds).values
skew(dt, nan_policy = 'omit')
```

This is quite high! Next, T2.

```python
times_T2 = data.loc[data.T2_missing==0]['Time']
dt = times_T2.diff().apply(lambda x: x.seconds).values
skew(dt, nan_policy = 'omit')
```

Skewness values are super high. It looks like imputation is going to be painful.

We can also look at the correlations between the various time columns and the missingness index for T1 and T2 to see if we can find some correlation that might explain the missing data.

```python
corrs = data[['T1_missing', 'T2_missing', 'Year', 'Month', 'Quarter', 'DoW', 'Hour']].corr(method='spearman')
sns.heatmap(corrs, annot=True)
plt.savefig('images/heatmap.png')
```

Spearman rank correlation identifies monotic relationships between variables. The only significant relationship in the matrix is the correlation between the missingness of T2 and the day of the week. We will explore this further below.

Hopefully most of the missing data could be explained by some sort of a holiday period explanation. Let's try correcting for that. We will exclude all of the weekend data and assume we live in a 5-day per week universe. We will still have to perform some form of imputation for the missing values that remain.

We can do a day-hour analysis to figure out if weekends have any impact on the missingness of the data. Let's look at T1 first.

```python
sns.heatmap(pd.pivot_table(data, values = 'T1_missing', index = 'Hour', columns = 'DoW', aggfunc = np.sum)/pd.pivot_table(data, values = 'T1_missing', index = 'Hour', columns = 'DoW', aggfunc = np.size))
```

Take a look at T2

```python
sns.heatmap(pd.pivot_table(data, values = 'T2_missing', index = 'Hour', columns = 'DoW', aggfunc = np.sum)/pd.pivot_table(data, values = 'T2_missing', index = 'Hour', columns = 'DoW', aggfunc = np.size))
```

Both variables are entirely empty on Saturdays and all the way until 9pm on Sundays. T2 suggests that there is no data after 9pm on Friday as well. T1 does have some data after 9pm but the missing percentage is quite high relative to all the other hour slots of all other days.

This indicates that we might be looking at a system with a 9pm to 9pm trading day, with a weekend spanning 9pm Friday to 9pm Sunday.

Another pattern that is visible especially in T1 is that there is more missing data towards the end of the day near the 7, 8, 9pm time blocks.

Why does the 9pm and 10pm blocks of T1 have data? Let's start by taking a look at the data in the Friday 10pm block.

```python
data_friday_10pm = data.loc[(data['DoW'] == 4) & (data['Hour'] == 22) & (data['T2_missing'] == 0)][['Time', 'T2']]
data_friday_10pm
```

It appears the datapoints are from exactly 10pm for each of the days. These can probably be safely disregarded from analysis if needed.

How about the 9pm blocks?

```python
data_friday_9pm = data.loc[(data['DoW'] == 4) & (data['Hour'] == 21) & (data['T2_missing'] == 0)][['Timedate', 'Time']]
data_friday_9pm.groupby('Timedate').count()
```

This means there are 255 Fridays with 9pm blocks with valid data in it. How many Fridays are there in th entire time period we have data?

```python
np.arange('2008-01-04', '2013-01', np.timedelta64(7, 'D'), dtype='datetime64[D]').size
```

Out of all 261 Fridays in the data range, 255 Fridays have a Friday 9pm block measurement with various numbers of measurements within the block.

This means we can probably safely drop that hour block and create a regular 9pm-9pm trading day setup with 5 trading days a week.

We will create a new column in the data table to allow us to keep track of the 9pm-9pm trading days. A Monday trading day will start on Sunday at 9pm and end on Monday at 9pm. The variable names will begin with an 'r' to indicate the renormalization of time.

```python
data['rt'] = data.t + 0.125
data['rTime'] = pd.to_datetime(data.rt-719529, unit='D')
data['rtdate'] = np.floor(data.rt).astype(int)
data['rTimedate'] = pd.to_datetime(data.rtdate-719529, unit='D')
data['rDoW'] = data.rTime.apply(lambda row:row.day_of_week)
data['rHour'] = data.rTime.apply(lambda row:row.hour)

data.head()
```

Now let's drop the weekend data from the analysis.

```python
rdata = data.copy()
rdata = rdata.loc[rdata['rDoW'] <5]
rdata.groupby('rDoW')['t'].count()
```

Now we have cleaned up all of the weekend noise, there obviously are missing data still remaning in the time series. How much of the data points are actually still missing and how many days are completely missing?

```python
print(f"Missing {rdata['T1_missing'].sum()} data points among {rdata['T1_missing'].count()} rows of variable T1 ({rdata['T1_missing'].sum()*100.0/rdata['T1_missing'].count():.2f}%)")
print(f"Missing {rdata['T2_missing'].sum()} data points among {rdata['T2_missing'].count()} rows of variable T2 ({rdata['T2_missing'].sum()*100.0/rdata['T2_missing'].count():.2f}%)")
```

Before imputations, let's take a quick peek at the frequency spectra of the missingness indices to see if there are any other obvious patterns.

Nothing really stands out in either plot other than the regular low frequency behaviours we might expect from most signals. T1 does seem to have some peaks at 20 and 60 intervals. There is also a small peak at 40. These might be some data missing due to a monthly occurrence.

To perform imputations, we will need to assume that the missing values are missing at random and there is no insidious pattern to them(otherwise we might as well stop the analysis now). With that assumption, this means that the overall distribution of data in T1 and T2 should not be affected by the missing data, and the complete dataset should also follow the same distribution as the incomplete one we have.

We will compute the first few moments of the distributions of T1 and T2, with and without imputations to compare between imputation methods. The moments calculation is suitable both because they parametrise the overall distributions reasonably in many cases, and they are impervious to data missing at random. We also include the KS test to see if the distribution is significantly altered by the imputation.

```python
imp_methods = [None, "bfill", "ffill", "linear"]
def get_stats(x):
    x = x.dropna()
    return np.array([x.mean(), x.std(), x.skew(), kstest(rdata['T1'].dropna(), x).pvalue])


stats_T1 = np.array(list(map(lambda x: get_stats(rdata['T1']) if (x is None) else get_stats(rdata['T1'].interpolate(method = x)), imp_methods)))
stats_T2 = np.array(list(map(lambda x: get_stats(rdata['T2']) if (x is None) else get_stats(rdata['T2'].interpolate(method = x)), imp_methods)))
```

```python
sns.heatmap(stats_T1, yticklabels = ['Raw data', 'Back Fill', 'Forward Fill', 'Linear Interpolation'], xticklabels=['Mean', 'SD', 'Skewness', 'KS statistic'])
```

```python
sns.heatmap(stats_T2, yticklabels = ['Raw data', 'Back Fill', 'Forward Fill', 'Linear Interpolation'], xticklabels=['Mean', 'SD', 'Skewness', 'KS statistic'])
```

All the imputation schemes look equally good(or equally bad). This is sort of expected as everything apart from linear imputation draws from the time series values themselves. Can we try another test to differentiate between them? 

How about the correlation between the two series? Perhaps we can see if the imputation has any effect on that?

```python
dfT1 = rdata['T1'].to_frame(name="T1")
for method in imp_methods[1:]:
    dfT1[method+'_T1'] = rdata['T1'].interpolate(method=method)
dfT2 = rdata['T2'].to_frame(name="T2")
for method in imp_methods[1:]:
    dfT2[method+'_T2'] = rdata['T2'].interpolate(method=method)
    
dfT = pd.concat([dfT1, dfT2], axis=1)
dfT.corr(method='spearman')
```

The correlation between raw T1 and T2 is 0.338768. All of the imputation methods seem to predict very similar values as well.

While there are many other methods to impute the data that we could try including multiple imputation which has become quite popular in recent years, in the interest of time, we will select forward filling as the imputation method( a backfill will still need to be performed on the very first of sequences of missing values in the two series). Another path that we could explore is to fit a model for the relationship between T1 and T2 and then using that relationship to impute missing values.

Next we must choose a discretisation. This will allow us to smooth out some of the data as well. We will set the smallest time scale of interest to daily frequencies. We will first create data at a daily frequency. In performing the downsampling, we could follow several methods. Close-to-close price, mean price, median price are all great options. But given that we performed quite a bit of imputation these may be quite biased as well. The close-to-close price may not be ideal since data tended to be missing more often towards the end of the trading day at 9pm.

To deal with this, I propose a hybrid approach. Given that almost all days have at least one data point available, we can use the median of the data within one day as the price value for that day. Choosing the median will also remove short-term volatility outliers. But on the downside, we won't be able to trade on intra-day variations.

Having done this for days with data, we perform forward filling and backfilling as needed for the rest of the missing days(this will only be needed for 25 days of the T1 series). We also choose to work in the log time scale as we are mostly interested in the variations of the values and not the values themselves.

```python
dT2 = np.log(rdata.groupby('rTimedate')['T2'].mean())
dT2.plot()
```

```python
dT1 = np.log(rdata.groupby('rTimedate')['T1'].mean())
dT1.fillna(method = 'ffill', inplace = True)
dT1.fillna(method = 'bfill', inplace = True)
dT1.plot()
plt.legend(['T1'])
```

Let's also compute log returns. In log space, the returns are just differences.

```python
dailyT1 = dT1.diff().shift(-1).dropna()
dailyT1.plot()
plt.legend(['T1'])
```

```python
dailyT2 = dT2.diff().shift(-1).dropna()
dailyT2.plot()
plt.legend(['T2'])
```

These log return transoformation not only gives us a variable of interest, but also gets rid of much of the non-stationary trends in most cases. However, there may still be longer term trends hiding in the data.

Note that we have completely disregarded price. We could possibly try calculating the trend using some sort of a smoothing method like a sliding window with Gaussian or Exponential smoothing. There are also other filters that we could use. But since the smoothers all act as low pass filters, they also tend to introduce artefacts of their own. Given the small signal to volatility scales we are dealing with, these artefacts could be dangerous and hence I will stick to the differencing to de-trend the series completely. That is not to say there may not be other techniques that will be useful when predicting the trend of data. This also means we will not be able to analyse pure prices but will always look only at price variations.

Just to see the effects of smoothing for finding trend, I used a Savitzky-Golay filter to try to find the trend of the log prices. The filter produces aesthetically pleasing results, but also seems to come with some artefacts that we will need to filter. Filtering these artefacts and preserving high frequency components of the volatility signal would be quite challenging.

```python
from scipy.signal import savgol_filter
fT1 = savgol_filter(dT1, window_length = 199, polyorder =5, mode='nearest')
plt.plot(fT1)
plt.plot(dT1.values)
```

The trend generated looks quite impressive. Let's look at the frequency domains of the signal with and without trend.

```python
ffT1 = np.fft.fft(dT1.values)
freqT1 = np.fft.fftfreq(ffT1.shape[-1])
plt.plot(np.abs(ffT1[1:100]))
```

```python
ffT1 = np.fft.fft(dT1.values - fT1)
freqT1 = np.fft.fftfreq(ffT1.shape[-1])
plt.plot(np.real(ffT1)[1:200])
```

There is an obvious amplification of one frequency within the de-trended signal, with respect to the rest of the frequency components.

Now we turns back to our log return series. We will first establish that the time series indeed have no trend and also check if the series show volatility clustering.

```python
adfuller(dailyT2.values)[1], dailyT2.mean()
```

Let's check if there is any volatility clustering(the plots sure did look like there was volatility clustering). We will examine the partial-autocorrelation of the log returns to check if there is evidence of autoregressive behaviour. The autocorrelations of the absolute or squared log returns should indicate if there is also any volatility clustering. Volatility clustering would imply heteroskedasticity, at least in a conditional sense.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_pacf(dT1, lags=100, title = 'Partial Autocorrelation of T1');
```

```python
plot_pacf(abs(dailyT1), lags=50, title = 'Partial Autocorrelation of abosolute value of T1');
```

```python
plot_acf(dailyT2, lags=50, title = 'Partial Autocorrelation of T2');
```

```python
plot_pacf(abs(dailyT2), lags=100, title = 'Partial Autocorrelation of abosolute value of T2');
```

Both series show very little autocorrelation beyond the first lag coefficient, indicating that an autoregressive order one model might explain the data well. They both also show some heteroskedasticity with multiple siginificant non zero lag autocorrelations of the absolute series.

Let's take a look at the distributions again.

Are they normal?

```python
print(f"Jarque-Bera p-value for T1 is:{jarque_bera(dailyT1).pvalue}")
print(f"Jarque-Bera p-value for T2 is:{jarque_bera(dailyT2).pvalue}")
```

Both are definitely not normal. This will be significant for our modelling later. Let's take a look at the distributions as well.

```python
def fit_norm(data):
    mu, std = norm.fit(data)
    plt.hist(data, bins=100, density=True, alpha=0.6, color='b')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)  
    plt.plot(x, p, 'k', linewidth=2)
    

fit_norm(dailyT1)
plt.title('T1 Normal fit')
```

```python
fit_norm(dailyT2)
plt.title('T2 Normal fit')
```

The distributions, while looking pretty symmetric, are way more peaked than the corresponding normal distribution. The Jarque-Bera test also confirms this.

Let's take a look at their statistics.

```python
statsT1 = [dailyT1.mean(), dailyT1.std(), dailyT1.skew(), dailyT1.kurtosis()]
statsT1
```

```python
statsT2 = [dailyT2.mean(), dailyT2.std(), dailyT2.skew(), dailyT2.kurtosis()]
statsT2
```

Yup, the skewness and kurtosis confirms what we see visually above. Both distributions are leptokurtic with high peaks and heavy tails. T1 has some right skew.

How would a Student-T distribution fit this?

```python
def fit_t(data):
    df, mu, std = tdist.fit(data, loc = data.mean(), scale = data.std())
    print(df, mu, std)
    plt.hist(data, bins=100, density=True, alpha=0.6, color='b')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = tdist.pdf(x, df, mu, std)  
    plt.plot(x, p, 'k', linewidth=2)
    return df, mu, std
    
df, mu, std = fit_t(dailyT1)
plt.title('T1 T-distribution fit')
```

```python
kstest(dailyT1, lambda x: tdist.cdf(x, df = df, loc = mu, scale = std))
```

```python
df, mu, std = fit_t(dailyT2)
plt.title('T2 T-distribution fit')
```

```python
kstest(dailyT2, lambda x: tdist.cdf(x, df = df, loc = mu, scale = std))
```

This look great! Both distributions fit great to the eye and T2 even seems to pass the KS test. T1 is probably better approximated by a Laplacian or some other peaked distribution. Let's try the Laplacian on T1.

```python
def fit_laplace(data):
    mu, std = laplace.fit(data, loc = data.mean(), scale = data.std())
    print(mu, std)
    plt.hist(data, bins=100, density=True, alpha=0.6, color='b')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = laplace.pdf(x, mu, std)  
    plt.plot(x, p, 'k', linewidth=2)
    return mu, std
mu, std = fit_laplace(dailyT1)
plt.title('T1 Laplace fit')
```

```python
kstest(dailyT1, lambda x: laplace.cdf(x, loc = mu, scale = std))
```

It seems the Laplacian doesn't do much too better of a job than the t distribution. So we will stick with the t distribution for now.
